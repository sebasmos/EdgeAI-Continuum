```python
import os
import pandas as pd

directories = [d for d in os.listdir(results_path) if os.path.isdir(os.path.join(results_path, d))]
print(directories)
dfs = []
for directory in directories:
    file_path = os.path.join(results_path, directory, 'metrics_combined.csv')
    if os.path.exists(file_path):
        df = pd.read_csv(file_path)
        df['folder'] = directory  
        dfs.append(df)

if dfs:
    all_data = pd.concat(dfs, ignore_index=True)
    
    aggregated = all_data.groupby('metric').agg(['mean', 'std']).reset_index()
    
    aggregated.columns = ['_'.join(col).strip('_') for col in aggregated.columns.values]
    
    aggregated.rename(columns={'metric': 'metric_name'}, inplace=True)
    aggregated.set_index('metric_name', inplace=True)  
    aggregated.to_csv(f'{name}_results.csv')
else:
    print("No metrics_combined.csv files found in the specified results_path.")
aggregated
```

# dataClay - Client results

```python
import os
import glob
import pandas as pd

# exp3: mac as server, orangepi as client

# folder = "/Users/sebasmos/Desktop/ucd_fun_horizon_icos_papers/Dataclay_paper/T1 - DataClay/T1_Local_DataClay_Client_mac2orangepi"
# output_file =  "T1 - Local (DataClay Client)_mac2orangepi.csv"
folder = "/Users/sebasmos/Desktop/ucd_fun_horizon_icos_papers/Dataclay_paper/AIoffload/results/Mac-OrangePI_client"
output_file="Mac-OrangePI_client.csv"

# # Ryzen-Mac_client
# folder = "/Users/sebasmos/Desktop/ucd_fun_horizon_icos_papers/Dataclay_paper/AIoffload/results/Ryzen-Mac_client"
# output_file="Ryzen-Mac_client.csv"

# # Ryzen-OrangePI_client
# folder = "/Users/sebasmos/Desktop/ucd_fun_horizon_icos_papers/Dataclay_paper/AIoffload/results/Ryzen-OrangePI_client"
# output_file="Ryzen-OrangePI_client.csv"

csv_files = glob.glob(os.path.join(folder, "*.csv"))
dataframes = []
for file in csv_files:
    df = pd.read_csv(file, comment='/')
    dataframes.append(df)

if dataframes:
    combined_df = pd.concat(dataframes, ignore_index=True)
else:
    combined_df = pd.DataFrame()

results = {}
if "memory" in combined_df.columns:
    results["memory_mean"] = combined_df["memory"].mean()
    results["memory_std"] = combined_df["memory"].std()
else:
    results["memory_mean"] = None
    results["memory_std"] = None

if "total proc. (s)" in combined_df.columns:
    results["total_proc_mean"] = combined_df["total proc. (s)"].mean()
    results["total_proc_std"] = combined_df["total proc. (s)"].std()
else:
    results["total_proc_mean"] = None
    results["total_proc_std"] = None

results_df = pd.DataFrame([results])
results_df.to_csv(output_file, index=False)
print(f"Combined results saved to {output_file}")
```
